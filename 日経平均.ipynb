```python
import yfinance as yf

code = "^N225"  # 日経225

nikkei = yf.Ticker(code)

# "1mo"で1ヵ月分取得
df_nikkei = nikkei.history(period="1mo")
```

```python
df_nikkei
```

```python
from matplotlib import pyplot as plt
df_nikkei['Open'].plot(kind='line', figsize=(8, 4), title='Open')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import numpy as np

df_nikkei['Open'].plot(kind='line', figsize=(8, 4), title='Open')

# 最大値を取得
xmax = df_nikkei['Open'].idxmax()
ymax = df_nikkei['Open'].max()

# annotation を追加、矢印の設定を追加
plt.annotate(f'最大値: {ymax:.2f}',
             xy=(xmax, ymax), xytext=(xmax, ymax + 100),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))  # 矢印の設定

plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```
